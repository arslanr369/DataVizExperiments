## Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Create Visualizations

The data is visualized using various charts and plots to present the information in a clear and visually appealing manner, allowing for easy interpretation and analysis of the budget allocation, revenue targets, foreign assistance, and expenditure categories.

## DataFrame from the Provided Data

```python
data = {
    'Federal Budget 2023-24': ['Total Budget', 'FBR Tax collection', 'Provinces', 'Defence Budget', 'Interests and loans'],
    'Amount in billion': [14460, 9200, 5276, 1804, 7303]
}
df = pd.DataFrame(data)

# Bar plot for Federal Budget
plt.figure(figsize=(10, 6))
sns.barplot(x='Federal Budget 2023-24', y='Amount in billion', data=df)
plt.xlabel('Category')
plt.ylabel('Amount in billion')
plt.title('Federal Budget 2023-24')

# Add value labels to the bars
for index, row in df.iterrows():
    plt.text(index, row['Amount in billion'], str(row['Amount in billion']), ha='center', va='bottom')

plt.show()
```

## DataFrame for Revenue Targets

DataFrame for Revenue Targets represents the revenue targets and corresponding amounts for various categories, providing insights into the projected sources of income for the given period.

```python
revenue_targets_data = {
    'Revenue Targets': [
        'Total Revenue Target', 'Non Tax Revenue', 'Direct taxes', 'In direct taxes',
        'Income Tax', 'Customs duty', 'Workers Welfare fund', 'Capital Value',
        'Sales Tax', 'FED', 'Petroleum Levy', 'Defence services',
        'State Bank of Pakistan', 'Govt departments', 'Public corporations',
        'Public companies shares'
    ],
    'Revenue in Billion': [
        9200, 2943, 3759, 5441, 3713, 1178, 13.84, 81.7, 3578,
        725, 869, 41, 1113, 398, 118, 121
    ]
}
df_revenue_targets = pd.DataFrame(revenue_targets_data)

# Bar plot for Revenue Targets using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Revenue Targets', y='Revenue in Billion', data=df_revenue_targets, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Revenue in Billion')
plt.title('Revenue Targets')
plt.xticks(rotation=90)

# Add value labels to the bars
for index, row in df_revenue_targets.iterrows():
    plt.text(index, row['Revenue in Billion'], str(row['Revenue in Billion']), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## DataFrame for Possible foreign assistance

DataFrame for Possible foreign assistance displays the different categories and corresponding amounts of potential foreign assistance, providing an overview of potential sources of financial support from external entities.

```python
foreign_assistance_data = {
    'Possible foreign assistance': [
        'Total', 'Power Bond ', 'Loan from Saudi Arabia',
        'Islami Taraqiati Bank', 'Deposits from KSA', 'Deposits from UAE', 'Loan from China',
        'China CB loans', 'IMF', 'oil payments'
    ],
    'Revenue in Billion': [
        6971, 435, 870, 145, 580, 290, 1160, 1305, 696, 29.58
    ]
}
df_foreign_assistance = pd.DataFrame(foreign_assistance_data)

# Bar plot for Possible foreign assistance using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Possible foreign assistance', y='Revenue in Billion', data=df_foreign_assistance, palette='rocket')
plt.xlabel('Category')
plt.ylabel('Revenue in Billion')
plt.title('Possible foreign assistance')
plt.xticks(rotation=90)

# Add value labels to the bars
for index, row in df_foreign_assistance.iterrows():
    plt.text(index, row['Revenue in Billion'], str(row['Revenue in Billion']), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

##  DataFrame for Category and Amount

DataFrame for Category and Amount represents different categories of revenue and their corresponding amounts, providing insights into the distribution and composition of revenue sources.

```python
category_amount_data = {
    'Category': [
        'Tax Revenue', 'Non-Tax Revenue',
        'Gross Revenue', 'Less Provincial Share',
        'Net revenue', 'Non Bank Borrowing',
        'Net External',
        'Bank Borrowing',
    ],
    'Amount': [
        9200, 2963, 12163, 5276, 6887, 1906, 2527, 3124,
    ]
}
df_category_amount = pd.DataFrame(category_amount_data)

# Pie chart for Category and Amount using Plotly
fig = px.pie(df_category_amount, values='Amount', names='Category', 
             title='Category and Amount', hole=0.4,
             color_discrete_sequence=px.colors.qualitative.Set3)

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(showlegend=False, font=dict(size=14))

fig.update_layout(height=600, width=800)

fig.show()
```

## DataFrame for Total Resources

DataFrame for Total Resources shows the allocation and amounts of various revenue sources, providing an overview of the total available resources for financial planning and decision-making.

```python
total_resources_data = {
    'Total Resources': [
        'Tax Revenue (FBR)',
        'Non-Tax Revenue',
        'Gross Revenue Receipts',
        'Less Provincial Share',
        'Net revenue Receipts',
        'Non Bank Borrowing',
        'Net External Receipts',
        'Bank Borrowing',
    ],
    'Amount in Billion': [
        9200, 2963, 12163, 5276, 6887, 1906, 2527, 3124,
    ]
}
df_total_resources = pd.DataFrame(total_resources_data)

# Bar plot for Total Resources with amounts shown on bars
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Resources', y='Amount in Billion', data=df_total_resources, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Amount in Billion')
plt.title('Total Resources')

# Add amount labels on top of each bar
for i, value in enumerate(df_total_resources['Amount in Billion']):
    plt.text(i, value, str(value), ha='center', va='bottom', fontweight='bold')

plt.xticks(rotation=90)
plt.show()
```

## DataFrame for Expenditure

DataFrame for Expenditure presents the distribution and amounts of different expenditure categories, offering insights into the allocation of funds and expenditure priorities.

```python
expenditure_data = {
    'Expenditure': [
        'Current', 'Interest Payments', 'Pension', 'Defence',
        'Grants and Transfers', 'Subsidies',
        'Civil Government', 'Emergency and others'
    ],
    'Amount': [
        13320, 7303, 761, 1804, 1464, 1074, 714, 200
    ]
}
df_expenditure = pd.DataFrame(expenditure_data)

# Define custom RGB colors
colors = [(0, 94, 157), (0, 143, 187), (42, 184, 197), (94, 205, 211),
          (144, 226, 213), (192, 239, 211), (231, 247, 201), (255, 255, 204)]
colors = [(r/255, g/255, b/255) for r, g, b in colors]

# Bar plot for Expenditure with custom colors
plt.figure(figsize=(10, 6))
bars = plt.bar(df_expenditure['Expenditure'], df_expenditure['Amount'], color=colors)
plt.xlabel('Category')
plt.ylabel('Amount in Billion')
plt.title('Expenditure')
plt.xticks(rotation=90)

# Add data labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, int(height),
             ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.show()
```

The data provides insights into the budget allocation, revenue targets, foreign assistance, and expenditure categories, aiding in informed decision-making and financial management for the year 2023-24.
